# To remove outliers of the dataset

```python
columns = [i for i in df]
columns
```

```python
try:
    for i in columns:
        plt.title(i)
        sns.boxplot(data=df[i])
        plt.show()
except:
    print('Categorical variable')
```

```python
q1=df.SepalWidthCm.quantile(0.25)
q3=df.SepalWidthCm.quantile(0.75)
iqr=q3-q1
iqr=1.5*iqr
LL=q1-iqr
Ul=q3+iqr
    
```

```python
data = df[(df.SepalWidthCm >LL) & (df.SepalWidthCm <Ul)]
```

```python
sns.boxplot(data=data['SepalWidthCm'])
plt.show()
```

All the outliers are now removed from the datset using the boxplot

# Modeling

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
```

```python
data['Species']=encoder.fit_transform(data['Species'])
data
```

Now all the target variables are converted to vectors so the machine can easily understand the datset 

# Splitting the dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x=data.drop(['Species'], axis=1)
y =data['Species']
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x_train
```

```python
y_train
```

By splitting the datset 70% is used for training the dataset and remaining 30% is used for testing. 

# Fitting into the model

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
[[x_test]]
```

# Accuracy Score

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred=knn.predict(x_test)
y_pred
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

# classification_report and confusion_matrix

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test, y_pred))
print("-----------------")
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn import metrics
print("MAR:",metrics.mean_absolute_error(y_pred,y_test))
print("RMSE:",np.sqrt(metrics.mean_squared_error(y_pred,y_test)))
print("R-Squared",metrics.r2_score(y_pred,y_test))
```

# Saving the Model

```python
import pickle
```

```python
saved_model=pickle.dumps(knn)
knn_from_pickle=pickle.loads(saved_model)
knn_from_pickle.predict(x_test)
```

Firstly we will be using the dump() function to save the model using pickle.Then we will be loading that saved model.lastly, after loading that model we will use this to make predictions.
